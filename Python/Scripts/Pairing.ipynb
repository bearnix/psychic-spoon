```python
df = pd.DataFrame(columns=[0, 1, 2])
sample = 100000
while sample > 0:
    sample = sample-1
    # make a list of As and Bs, all of even numbers, 13 men
    group = np.concatenate([np.full(37, 0), np.full(13, 1)])
    test =  []
    test2 = []
    while len(group)>0:
        random_index = np.random.randint(0, len(group))
        pair_v = group[random_index]
        group = np.delete(group, random_index)
        random_index = np.random.randint(0, len(group))
        pair_v += group[random_index]
        group = np.delete(group,random_index)
        test.append(pair_v)
    #unique, counts = numpy.unique(test, return_counts=True)
    counter = collections.Counter(test)
    #print(counter)
    df.loc[len(df)]=counter
#print(df)
```
